```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

# Part 1

```python
def lorenz(t, W, sigma=10.0, r=28.0, b=8/3):
    """
    Compute the time derivatives of the Lorenz equations, which are defined by the following coupled ODEs:

        Xdot = -sigma (X - Y)
        Ydot = rX - Y - XZ       # nonlinear
        Zdot = XY - bZ           # nonlinear

    Parameters:
    W : array-like of shape (3,)
        Current state vector [X, Y, Z]
    
    sigma : float
            Prandtl number (ratio of the kinematic viscosity to the thermal diffusivity)
            Pre-set value (from Lorenz' paper) is 10.0
        
    r : float
        Rayleigh number (depends on vertical temperature gradient of atmosphere)
        Pre-set value (from Lorenz' paper) is 28.0
        
    b : float
        Dimensionless length scale
        Pre-set value (from Lorenz' paper) is 8/3

    Returns:
    Wdot : ndarray of shape (3,)
           Time derivatives [Xdot, Ydot, Zdot]
    
    """
    X, Y, Z = W

    Xdot = -sigma*(X - Y)
    Ydot = r*X - Y - X*Z
    Zdot = X*Y - b*Z

    return [Xdot, Ydot, Zdot]
```

# Part 2

```python
# initial conditions
W0 = [0.0, 1.0, 0.0]

# time interval
t0 = 0.0 # initial 
tf = 60.0 # final

# linspace of times to evaluate at 
t_eval = np.linspace(t0, tf, 3000)

# solving ODE
# default is 4th order runge kutta
sol = solve_ivp(lorenz, (t0, tf), W0, t_eval=t_eval, dense_output = True)

# extracting solutions
X = sol.y[0]
Y = sol.y[1]
Z = sol.y[2]

t = sol.t

print(X, Y, Z)
```

# Part 3: Lorenz' Fig. 1

```python
# Lorenz used N = t/dt with dt = 0.01
N = t / 0.01

plt.figure(figsize=(10, 7))
plt.plot(N, Y)

plt.xlabel("N = t/dt")
plt.ylabel("Y")
plt.title("Lorenz Figure 1")

plt.grid(True)
plt.show()
```

# Part 4: Lorenz' Fig. 2

```python
# fig 2 phase space projections
t_2 = np.linspace(14, 19, 10000)

W_2 = sol.sol(t_2)

X_2 = W_2[0]
Y_2 = W_2[1]
Z_2 = W_2[2]

# X-Y plane
plt.figure(figsize=(10, 7))
plt.plot(Y_2, X_2)

plt.xlabel("Y")
plt.ylabel("X")
plt.title("Lorenz Figure 2: X-Y Phase Space")

plt.grid(True)
plt.show()


# Y-Z plane
plt.figure(figsize=(10, 7))
plt.plot(Y_2, Z_2)

plt.xlabel("Y")
plt.ylabel("Z")
plt.title("Lorenz Figure 2: Y-Z Phase Space")

plt.grid(True)
plt.show()
```

# Part 5: Initial Condition Sensitivity

```python
# slightly perturbed initial condition
W0_prime = np.array(W0) + np.array([0.0, 1e-8, 0.0]) # to add lists

# solving the perturbed system
sol_prime = solve_ivp(lorenz, (t0, tf), W0_prime, t_eval=t_eval)

# perturbed solution
X_prime = sol_prime.y[0]
Y_prime = sol_prime.y[1]
Z_prime = sol_prime.y[2]

# computing distance between trajectories
dist = np.sqrt((X_prime - X)**2 + (Y_prime - Y)**2 + (Z_prime - Z)**2)

# preventing log(0) errors 
dist[dist == 0] = 1e-16

# semilog plot
plt.figure(figsize=(10, 7))
plt.semilogy(t, dist)

plt.xlabel("Time")
plt.ylabel("Distance between trajectories")
plt.title("Divergence of Nearby Trajectories")

plt.grid(True)
plt.show()
```
