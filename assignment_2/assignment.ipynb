# Part 1: defining f(x), f'(x)

```python
def f(x):
    """ This function calculates the function f(x).
    
    Parameters: 
    x : float, unitless 
    
    Returns: 
    f(x) : float, unitless
           the function f evaluated at x
    """
    return x**3 - x**2 - 1
```

```python
def df(x):
    """This function calculates the derivative of the function f(x), with respect to x.
    
    Parameters: 
    x : float, unitless 
    
    Returns: 
    f'(x) : float, unitless 
            the derivative df/dx, evaluated at x,
    """
    return 3*x**2 - 2*x
```

# Part 2

```python
def newton(f, df, x0, epsilon=1e-6, max_iter=30):
    """
    Performs a Newton Iteration of the function f with derivative df
    Newton iteration determines the root x_n such that f(x_n) = 0.
    
    Parameters:
    f : callable
        function whose root is being determined
    df : callable
         derivative of function f 
    x0 : float
         initial root guess 
    epsilon : float, optional, default is 1e-6
              tolerance, chosen to indicate convergence
    max_iter: int, optional, default is 30
              number of iterations function will loop through to check for convergence

    Returns:
    float corresponding to root if root is found
    None else
    """
    x_n = x0

    for it in range(1, max_iter + 1):  # iterate through loop at most max_iter times 
        if abs(f(x_n)) < epsilon:  # checks for convergence
            print(f"Found root in {it} iterations")
            return x_n
        
        if df(x_n) == 0: # prevent div by 0
            print("Iteration failed") 
            return None
        
        x_n -= f(x_n) / df(x_n) # newtonian iteration

    # if max_iter exceeded:
    print("Iteration failed")
    return None
```
