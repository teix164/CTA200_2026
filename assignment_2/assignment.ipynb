# Part 1: defining f(x), f'(x)

```python
def f(x):
    """ This function calculates the function f(x)
    Parameters: x ------- array-like, the variable x, unitless 
    Returns: f(x) -------- array-like, the quantity f(x), unitless
    CHECK UR STRING NOTATION
    should mention x must be integer or float, not string 
    """
    return x**3 - x**2 - 1
```

```python
def df(x):
    """This function calculates the derivative of the function f(x), with respect to x
    Parameters: x --------- array-like, the variable x, unitless
    Returns: f'(x) -------- array-like, the quantity df/dx 
    """
    return 3*x**2 - 2*x
```

```python
def newton(f, df, x0, epsilon=1e-6, max_iter=30):
    """Performs a Newton Iteration of the function f with derivative df
    Newton iteration determines the root x_n such that f(x_n) = 0"""
    x_n = x0
    it = 0 
    f = f(x0)
    for abs(f) < epsilon and it <= max_iter:
        x_n -= f(x_n)/df(x_n)
        it += 1
        f = f(x_n)
    if abs(f) < epsilon:
        print(f"Found root in {it} iterations")
        return x_n
    else: 
        print("Iteration failed")
        return None
        
```
