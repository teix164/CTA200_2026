```python
def newton(f, df, x0, epsilon=1e-6, max_iter=30):
    """
    Performs a Newton Iteration of the function f with derivative df
    Newton iteration determines the root x_n such that f(x_n) = 0
    Parameters
    f : function 
    df : derivative of function f 
    x0 : initial root guess
    epsilon :
    max_iter: number of iterations function will loop through to chekc for convergence

    Returns:
    float if root is found
    None else
    """
    x_n = x0

    for it in range(1, max_iter + 1):  # iterate at most max_iter times
        fx = f(x_n)
        if abs(fx) < epsilon:  # check convergence
            print(f"Found root in {it} iterations")
            return x_n
        dfx = df(x_n)
    
        x_n -= fx / (dfx + 1e-16) # prevent div by 0

    # if here, max_iter exceeded
    print("Iteration failed")
    return None
```
